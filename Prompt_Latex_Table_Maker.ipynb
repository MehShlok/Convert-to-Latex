```python


file = "table1.txt"
#Beginning
n = int(input("Enter No. of Columns: \n"))
m = int(input("Enter the No of Rows(Don't count the Headings Row): \n"))

with open(file, 'w') as f:
    f.write("\\begin {table}[htbp]\n")
    f.write("\t\centering\n")
    f.write("\t\\begin {tabular}{|l| * {%s}{|c|}}\n"%n)
    f.write("\t\hline\n")
    #print("\\begin {table}[htbp]\n")
    #print("\t\centering\n")
    #print("\t\\begin {tabular}{|l| * {%s}{|c|}}\n"%n)

    #Headings
    S=["" for x in range(n)]
    i=0
    for i in range(n-1):
        S[i] = input("Enter Heading %s:"%(i+1))
        if(i==0):
            S[i] = "\t\\textbf{" + S[0] + "}"
        else:
            S[i] =  S[i-1] + " & " + "\\textbf{" + S[i] + "}"

    S[n-1] = S[n-2] + " & " + "\\textbf{" + input("Enter the last heading") + "}" + "\\" + "\\"

    f.write(S[n-1]+"\n")
    #print(S[n-1])

    #Content
    for j in range(m):
        print("Enter Values in row %s:"%(j+1))
        L=["" for y in range(n)]
        for k in range(n-1):
            if(k==0):
                L[k] = "\t" + input("Enter the value under Heading %s:"%(k+1))
            else:
                L[k] = L[k-1] + " & " + input("Enter the value under Heading %s:"%(k+1))

        L[n-1] = L[n-2] + " & " + input("Enter the last value of this row :") + " \\" + "\\"

        f.write(L[n-1]+"\n")
        #print(L[n-1])

    #The End
    f.write("\t\hline\n")
    f.write("\t\end{tabular}\n")
    #print("\t\end{tabular}\n")
    c = input("Enter caption for your table:")
    f.write("\t\caption{%s}\n"%c)
    #print("\t\caption{%s}"%c)
    f.write("\end{table}\n")
    #print("\end{table}")

f.close()




```
